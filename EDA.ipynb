# CHecking details of the country with the highest score

```python
maxss = data[data["Score"] == data["Score"].max()]
maxss = pd.DataFrame(maxss)
maxss
```

# Checking the details of the country with the lowest score

```python
minss = data[data["Score"] == data["Score"].min()]
minss = pd.DataFrame(minss)
minss
```

# Dropping unecessary column for analysing data

```python
data1 = data.drop(["Overall rank", "Country or region"], axis=1)
```

# Plotting heatmap of dataframe 

```python
corelation = data1.corr()
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True )
```

# Plotting histogram for "Score"

```python
data1["Score"].plot(kind='hist', edgecolor='black')
```

# Plotting histogram for "GDP per capita"


```python
data1["GDP per capita"].plot(kind='hist', edgecolor='black')
```

# Plotting pairplot of the dataframe

```python
sns.pairplot(data1)
```

# Plotting relation plot between "Healthy life expectancy" and "Freedom to make life choices"

```python
sns.relplot(x=data1["Healthy life expectancy"], y=data1["Freedom to make life choices"])
```

# Plotting relation plot between "Social support" and "Generosity"

```python
sns.relplot(x=data1["Social support"], y=data1["Generosity"])
```

# Plotting distribution plot to analyse the variation in data distribution

```python
sns.distplot(data1["Score"])
```
